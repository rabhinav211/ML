Diabets.csv

```python
import pandas as pd

data = pd.read_csv('diabetsdataset.csv')
print("Sample data:")
print(data.head())
```

```python
missing_values = data.isnull().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])
```

```python
#filling missing values
from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(strategy="median")
imputer2 = SimpleImputer(strategy="mean")

df_copy=data

# Step 2: Fit the imputer on the "Age" and "Salary"column
# Note: SimpleImputer expects a 2D array, so we reshape the column
imputer1.fit(df_copy[["AGE"]])
imputer2.fit(df_copy[["TG"]])

# Step 3: Transform (fill) the missing values in the "Age" and "Salary"c column
df_copy["AGE"] = imputer1.transform(data[["AGE"]])
df_copy["TG"] = imputer2.transform(data[["TG"]])

# Verify that there are no missing values left
print(df_copy["AGE"].isnull().sum())
print(df_copy["TG"].isnull().sum())
```

```python
#encoding
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

# Fit and transform the 'Gender' column
df_copy['Gender'] = encoder.fit_transform(df_copy[['Gender']])

# Print the updated DataFrame to verify the encoding
print(df_copy.head())
```

```python
#Min-Max Scaling
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
normalizer = MinMaxScaler()
df_copy[['No_Pation']] = normalizer.fit_transform(df_copy[['No_Pation']])
df_copy.head()
```

```python
#Standardization
scaler = StandardScaler()
df_copy[['AGE']] = scaler.fit_transform(df_copy[['AGE']])
df_copy.head()
```

```python
#outliers
import numpy as np

df_encoded_copy1=df_copy
df_encoded_copy2=df_copy
df_encoded_copy3=df_copy
#IQR
Q1 = df_encoded_copy1['No_Pation'].quantile(0.25)
Q3 = df_encoded_copy1['No_Pation'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_encoded_copy1['No_Pation'] = np.where(df_encoded_copy1['No_Pation'] > upper_bound, upper_bound,
                        np.where(df_encoded_copy1['No_Pation'] < lower_bound, lower_bound, df_encoded_copy1['No_Pation']))

print(df_encoded_copy1.head())
```

```python
#zscore
from scipy import stats
df_encoded_copy2['No_Pation_zscore'] = stats.zscore(df_encoded_copy2['No_Pation'])
df_encoded_copy2['No_Pation'] = np.where(df_encoded_copy2['No_Pation_zscore'].abs() > 3, np.nan, df_encoded_copy2['No_Pation'])  # Replace outliers with NaN
print(df_encoded_copy2.head())
```
